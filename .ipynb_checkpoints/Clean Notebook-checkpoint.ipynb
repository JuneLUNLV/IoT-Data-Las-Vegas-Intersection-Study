```python
import pandas as pd
```

```python
df = pd.read_csv("intersection.csv")
```

```python
new_df = df[['OBJECTID', 'Name', 'LAT_DECIMAL', 'LON_DECIMAL', 'AADT_2007',
       'AADT_2008', 'AADT_2009', 'AADT_2010', 'AADT_2011', 'AADT_2012',
       'AADT_2013', 'AADT_2014', 'AADT_2015', 'AADT_2016', 'AADT_2017',
       'AADT_2018', 'AADT_2019', 'AADT_2020']]
```

```python
new_df
```

```python
str_columns = ['AADT_2007',
       'AADT_2008', 'AADT_2009', 'AADT_2010', 'AADT_2011', 'AADT_2012',
       'AADT_2013', 'AADT_2014', 'AADT_2015', 'AADT_2016', 'AADT_2017',
       'AADT_2018', 'AADT_2019', 'AADT_2020']
for column in str_columns:
    new_df[column] = new_df[column].str.replace(",","")
```

```python
for column in str_columns:
    new_df[column] = pd.to_numeric(new_df[column])
```

```python
for column in str_columns:
    new_df[column].fillna(float(new_df[column].mean()), inplace=True)
```

```python
new_df
```

```python
new_df.to_csv("cleaned_intersection_fillna_column_mean.csv")
```
